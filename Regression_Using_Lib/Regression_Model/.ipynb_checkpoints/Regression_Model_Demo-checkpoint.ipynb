# Predict Boston House Value

Housing Values in Suburbs of Boston
The medv variable is the target variable.

Data description
The Boston data frame has 506 rows and 14 columns.

This data frame contains the following columns:

1. crim :per capita crime rate by town.

2. zn : proportion of residential land zoned for lots over 25,000 sq.ft.

3. indus : proportion of non-retail business acres per town.

4. chas : Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

5. nox :nitrogen oxides concentration (parts per 10 million).

6. rm : average number of rooms per dwelling.

7. age proportion of owner-occupied units built prior to 1940.

8. dis :weighted mean of distances to five Boston employment centres.

9. rad : index of accessibility to radial highways.

10. tax :full-value property-tax rate per $10,000.

11. ptratio pupil-teacher ratio by town.

12. black : 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

13. lstat lower status of the population (percent).

medv : median value of owner-occupied homes in $1000s.

```python
from sklearn.datasets import load_boston

dataset = load_boston()
X = pd.DataFrame(dataset["data"])
y = pd.DataFrame(dataset["target"])
y.columns = ["medv"]
features =  dataset["feature_names"]
X.columns  = features
X.head()
```

```python
X.dtypes
```

# Check Linear Relationship between Dependent & Independent Variable 

```python

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.concat([X,y],axis = 1)
corr = df.corr()
np.round(corr,2)
```

```python

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
sns.pairplot(df, hue='medv', size=2.5)
```

# Linear Regression Using StatsModel Libary

```python
import statsmodels.api as sm

X = sm.add_constant(X)
ols_model = sm.OLS(y,X).fit()
ols_model.summary()
```

```python
ols_predicted_X = pd.DataFrame(ols_model.predict(X))
ols_predicted_X = pd.concat([ols_predicted_X,y],axis = 1)
ols_predicted_X.columns = ["Predicted_Home_Price","Median_Home_Price"]
ols_predicted_X.plot()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
np.sqrt(mean_squared_error(y_pred= ols_predicted_X["Predicted_Home_Price"] , y_true= y))
```

# Check Assumptions of Linear Regression

```python
#2.Mean of the residual shoud be close to zero
res =  ols_predicted_X["Median_Home_Price"] - ols_predicted_X["Predicted_Home_Price"]
np.round(np.sum(res),2)
```

```python
#3. It should be homoscedasticity
import numpy as np
fitted_value = ols_predicted_X["Predicted_Home_Price"]
residual_value = ols_model.resid
plt.scatter(fitted_value,residual_value)
```

```python
#4.All the predictor variables & residuals should not be correlated
#X.drop("const",inplace = True , axis = 1)
df_cor = pd.concat([X,ols_model.resid],axis =1)
df_cor = df_cor.rename(columns={0: 'Residual'})
res_cor = df_cor.corr()
np.round(res_cor,2)
```

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(res_cor, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(res_cor, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
#5.Number of observations should be greater than number of independent variables .
obv = len(X)
col = len(X.columns)
obv,col,obv>col
```

```python
#6. No multicolinearity between predictor variables

#check vif for multicolinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["Features"] = X.columns
np.round(vif.sort_values(by = "VIF Factor"),2)
```
