```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

X, y = make_regression(n_samples = 300, n_features=1, noise=8, bias=2)
y2 = y**2
plt.scatter(X,y2)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y2)
pred_1 = model.predict(X)
plt.plot(X, model.predict(X))
plt.scatter(X, y2)
plt.title("Linear Model, Polynomial Degree = 1")
```

```python
from sklearn import metrics
import numpy as np

np.sqrt(metrics.mean_squared_error(y_pred= pred_1, y_true= y2))
```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures  
poly_features = PolynomialFeatures(degree = 3)  
X_poly = poly_features.fit_transform(X)
poly_model = LinearRegression()  
poly_model.fit(X_poly, y2)

pred = poly_model.predict(X_poly)
new_X, new_y = zip(*sorted(zip(X, pred))) # sort values for plotting
plt.plot(new_X, new_y)
plt.scatter(X,y2)
plt.title("Polynomial Degree = 3")
```

```python
from sklearn import metrics
import numpy as np

np.sqrt(metrics.mean_squared_error(y_pred= pred, y_true= y2))
```
